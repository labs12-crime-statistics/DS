```python
SEVERITY_HIGH = 100
SEVERITY_MEDIUM = 50
SEVERITY_LOW  = 20

def str_contains_any_substr(string, substrings):
    for subs in substrings:
        if subs in string:
            return True
    return False

havent_handled = []
def crime_severity(x):
    """
    Converts the primary type and description into a crime severity value.
    This value is based on how much a particular crime would affect a user
    that is either:
    1. Renting short-term and wants to roam around the city
    2. Renting long-term and wants to find a good neighborhood to roam in
    """
    description = x
    if ('VEH' in description):
        if 'THEFT FROM' in description or \
            'BURGLARY FROM' in description:
            return SEVERITY_LOW
        elif 'STOLEN' in description:
            return SEVERITY_MEDIUM
        elif 'THROWING' in description or \
             'SHOTS FIRED' in description:
            return SEVERITY_HIGH
              
    if 'THEFT' in description:
        if 'PETTY' in description or \
            'DEFRAUD' in description:
            return SEVERITY_LOW
        elif 'GRAND' in description:
            return SEVERITY_MEDIUM
        elif 'PERSON' in description:
            return SEVERITY_HIGH
        else:
            return SEVERITY_LOW
    
    if str_contains_any_substr(description,
                              ['BLOCKING', 'BIKE',
                               'UNAUTHORIZED COMPUTER ACCESS',
                               'BRIBERY',
                               'TRESPASSING', 
                               'OTHER MISCELLANEOUS CRIME',
                               'FALSE IMPRISONMENT',
                               'ORAL COPULATION',
                               'TILL TAP - PETTY ($950 & UNDER)',
                               'TILL TAP - ATTEMPT', 
                               'SHOPLIFTING - ATTEMPT',
                               'DOCUMENT WORTHLESS ($200.01 & OVER)',
                               'DOCUMENT WORTHLESS ($200 & UNDER)',
                               'CONTRIBUTING', 
                               'BIGAMY',
                              ]):
        return SEVERITY_LOW
    
    if str_contains_any_substr(description, 
                               ['PICKPOCKET', 'CREDIT CARDS',
                                'DRUNK ROLL', 'FORGERY',
                                'TELEPHONE', 'BURGLARY', 'ROBBERY',
                                'PURSE SNATCHING', 'BUNCO',
                                'INDECENT EXPOSURE', 
                                'CONTEMPT OF COURT', 
                                'COUNTERFEIT', 
                                'FAILURE TO YIELD',
                                'FAILURE TO DISPERSE',
                                'DRIVING WITHOUT OWNER CONSENT (DWOC)',
                                'CRUELTY TO ANIMALS',
                                'DISRUPT SCHOOL',                                
                               ]):
        return SEVERITY_MEDIUM
    
    if 'VANDALISM - FELONY' in description:
        return SEVERITY_MEDIUM
    elif 'VANDALISM - MISDEAMEANOR' in description:
        return SEVERITY_LOW
    
    if 'CHILD' in description:
        if 'ANNOYING' in description:
            return SEVERITY_LOW
        elif 'SIMPLE' in description:
            return SEVERITY_MEDIUM
        else:
            return SEVERITY_HIGH
        
    if 'SEX' in description:
        if 'INC MUTUAL CONSENT' in description:
            return SEVERITY_LOW
        else:
            return SEVERITY_HIGH
    
    if 'WEAPON' in description:
        if 'NO WEAPON DISPLAYED' in description:
            return SEVERITY_LOW
        else:
            return SEVERITY_MEDIUM
    
    if str_contains_any_substr(description,
                               ['DISCHARGE FIREARMS', 'SHOTS FIRED',
                                'FIREARMS RESTRAINING ORDER']):
        return SEVERITY_HIGH
    elif 'REPLICA FIREARMS' in description:
        return SEVERITY_MEDIUM
    
    if str_contains_any_substr(description,
                                ['RAPE', 'CHILD PORNOGRAPHY',
                                 'BATTERY', 'HUMAN TRAFFICKING',
                                 'KIDNAPPING', 'ASSAULT',
                                 'CRM AGNST CHLD',
                                 'ARSON', 
                                 'PEEPING TOM',
                                 'PROWLER', 
                                 'PIMPING',
                                 'VIOLATION OF COURT ORDER', 
                                 'BOMB SCARE',
                                 'STALKING',
                                 'DISTURBING THE PEACE',
                                 'LETTERS, LEWD', 
                                 'VIOLATION OF TEMPORARY RESTRAINING ORDER',
                                 'ILLEGAL DUMPING', 
                                 'RESISTING ARREST',
                                 'VIOLATION OF RESTRAINING ORDER', 
                                 'FALSE POLICE REPORT',
                                 'CRIMINAL HOMICIDE', 
                                 'EXTORTION', 
                                 'PANDERING',
                                 'THREATENING PHONE CALLS/LETTERS', 
                                 'LEWD CONDUCT', 
                                 'RECKLESS DRIVING', 
                                 'ABORTION/ILLEGAL',
                                 'BOAT - STOLEN', 
                                 'INCITING A RIOT', 
                                 'LYNCHING', 
                                 'LYNCHING - ATTEMPTED', 
                                 'CONSPIRACY',
                                 'DRUGS, TO A MINOR', 
                                 'MANSLAUGHTER, NEGLIGENT',
                                 'TRAIN WRECKING',
                                ]):
        return SEVERITY_HIGH
 
    havent_handled.append(description)
    return SEVERITY_LOW
```

```python
df['Severity'] = df['Crime Code Description'].apply(crime_severity)
```
