```python
def str_contains_any_substr(string, substrings):
    for subs in substrings:
        if subs in string:
            return True
    return False

other_categories = []
def location_category(x):
    location_description = x
    if (location_description == 'RESIDENCE') or \
        (location_description == 'SINGLE FAMILY DWELLING') or \
         str_contains_any_substr(location_description, ['HOUSE', 'RESIDENCE']):
        return 'INDOOR_RESIDENTIAL_HOME'

    if (location_description == 'APARTMENT') or \
        (location_description == 'MULTI-UNIT DWELLING (APARTMENT, DUPLEX, ETC)') or \
        str_contains_any_substr(location_description, ['CONDOMINIUM']):
        return 'INDOOR_RESIDENTIAL_APT'
    
    if str_contains_any_substr(location_description, ['SCHOOL']):
        return 'INDOOR_SCHOOL'
    
    if str_contains_any_substr(location_description, ['HOTEL', 'MOTEL']):
        return 'INDOOR_COMMERCIAL_HOTEL'

    if str_contains_any_substr(location_description, 
        ['STORE', 'RETAIL','RESTAURANT', 'BAR', 'TAVERN','CLUB',
         'NIGHTCLUB', 'GAS STATION', 'SHOP', 'SALON', 'MARKET',
         'BUSINESS', 'MART', 'MALL']):   
        return 'INDOOR_COMMERCIAL_STORE'

    if str_contains_any_substr(location_description, 
        ['BUILDING', 'OFFICE','INDUSTRIAL', 'FACILITY', 'CENTER']):   
        return 'INDOOR_COMMERCIAL_BUILDING_OFFICE_INDUSTRIAL'
    
    if (location_description == 'STREET'):
        return 'OUTDOOR_PUBLIC_STREET'
    
    if (location_description == 'SIDEWALK'):
        return 'OUTDOOR_PUBLIC_SIDEWALK'
    
    if (location_description == 'ALLEY'):
        return 'OUTDOOR_PUBLIC_ALLEY'
    
    if str_contains_any_substr(location_description, ['PARK', 'GARDEN']):
        return 'OUTDOOR_PUBLIC_OPENSPACE'
    
    if str_contains_any_substr(location_description, ['PARKING', 'LOT', 'GARAGE']):
        return 'OUTDOOR_PUBLIC_PARKING_LOT_GARAGE'
    
    if str_contains_any_substr(location_description, 
        ['TRAIN', 'BUS','SUBWAY', 'CTA', 'MTA']):   
        return 'OUTDOOR_PUBLIC_TRANSIT'
    
    if str_contains_any_substr(location_description, 
        ['YARD', 'PORCH','DRIVEWAY', 'BACKYARD']): 
        return 'OUTDOOR_PRIVATE_YARD'
    
    if str_contains_any_substr(location_description, ['VEHICLE']):
        return 'VEHICLE' 
    
    other_categories.append(location_description)
    return 'OTHER'
```

```python
df['Loc Category'] = df['Premise Description'].apply(location_category)
```
